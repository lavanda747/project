```python
fig = plt.figure(figsize=(20, 40))  # изобразим boxplot чтобы увидеть как распределяются наши объекты по признакам
for i, col in enumerate(num_cols, 1):
    ax = plt.subplot(4, 2, i)
    sns.boxplot(df[col])
    ax.set_title(col)
plt.show()
```

Возьмем за длину усов двойную длину ящика для каждого признака и посмотрим сколько объектов мы можем выкинуть. (не буду пока выкидывать выбросы)

```python
for col in num_cols:
    q_low = df[col].quantile(0.25)
    q_high = df[col].quantile(0.75)
    tentacle_length = q_high - q_low
    upper_tentacle = q_high + 2 * tentacle_length
    lower_tentacle = q_low - 2 * tentacle_length
    print(f' {col} - {len(df[(df[col] > lower_tentacle) & (df[col] < upper_tentacle)])}')
```

Отбросим выбросы в вещественных колонках и посмотрим на графики отношения цены к признаку

```python
for col in num_cols:
    q_low = df[col].quantile(0.25)
    q_high = df[col].quantile(0.75)
    tentacle_length = q_high - q_low
    upper_tentacle = q_high + 2 * tentacle_length
    lower_tentacle = q_low - 2 * tentacle_length
    df = df[(df[col] > lower_tentacle) & (df[col] < upper_tentacle)]
```

```python
plt.figure(figsize=(20,20))  # картина немного улучшилась и мы потеряли всего 1 процент от исходного количества объектов
for i, col in enumerate(df.columns[:-1], 1):
    ax = plt.subplot(4, 2, i)
    plt.bar(df[col], df['Rings'])
    plt.xlabel(col)
    plt.ylabel('Ring')
plt.show()
```

# Построить матрицу корреляций

Здесь мы можем построить матрицу корреляций только для вещественных признаков. Т.к у нас только один категориальный признак

```python
corr_df = df.corr()

sns.heatmap(corr_df, annot=True) 
```

Построим матрицу корреляций категориальных признаков к вещественным

```python
from scipy.stats import f_oneway
```

```python
Data = []

for c1 in df.columns:
    for c2 in df.columns:
        if df[c1].dtype == 'object' and df[c2].dtype != 'object':
            CategoryGroupLists = df.groupby(c1)[c2].apply(list)
            AnovaResults = f_oneway(*CategoryGroupLists)

            if AnovaResults[1] >= 0.05:
                Data.append({'Category' : c1, 'Numerical' : c2, 'Is correlated' : 'No'})
            else:
                Data.append({'Category' : c1, 'Numerical' : c2, 'Is correlated' : 'Yes'})

AnovaRes = pd.DataFrame.from_dict(Data)
AnovaRes  # видим что все вещественные признаки коррелируют с категориальным 
```
